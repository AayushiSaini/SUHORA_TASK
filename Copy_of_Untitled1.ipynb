```python
from PIL import Image
import matplotlib.pyplot as plt
import geopandas as gpd
import numpy as np
```

```python
img_path="/content/data/nepal_lc_2020.tif"
img=Image.open(img_path)
```

```python
plt.imshow(img)
plt.show()
```

```python
def change_band_combination(band1,band2,band3):
  new_img=img.split()
  new_img=Image.merge("RGB",(new_img[band1],new_img[band2],new_img[band3]))
  plt.imshow(new_img)
  plt.show()
```

```python
shp_path="/content/data/Nepal_Districts.shp"
shp_data=gpd.read_file(shp_path)
plt.figure(figsize=(100, 6))
plt.bar(shp_data['District'],height=shp_data['Area'])
plt.xlabel('District')
plt.ylabel('Area')
plt.title('Shapefile Attribute Data Chart')
plt.show()
```
