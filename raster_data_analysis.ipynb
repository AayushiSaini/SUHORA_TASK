```python
from rasterio.plot import show
show(ds,title='Land Cover Map of Nepal')
```

```python
with rasterio.open(r"/content/output/nepal_lc_2020_out.tif",'w',
                  driver=ds.driver,
                  height=ds.height,
                   width=ds.width,
                   count=ds.count,
                   crs=ds.crs,
                   transform=ds.transform,
                   dtype=data.dtype,
 ) as dst:
 dst.write(data)
```
